### Actualizar atributos de un shape

```python
import fiona
from fiona.transform import transform_geom

input_path = '/home/diego/git/AEPython_2024/data/islas_20170101_generalizada_etrs.shp'
output_path = '/home/diego/git/AEPython_2024/data/islas_20170101_generalizada_etrs_km.shp'

# Leer el shapefile original
with fiona.open(input_path) as shp:
    schema = shp.schema
    
    # Aadir un nuevo atributo al esquema
    schema['properties']['Area km2'] = 'float'

    # Crear un nuevo shapefile con el nuevo campo
    with fiona.open(output_path, 'w', driver='ESRI Shapefile', schema=schema, crs=shp.crs) as output:
        for feature in shp:
            # Calcular el nuevo valor basado en alguna operacin
            geometry = shape(feature['geometry'])
            areakm = geometry.area / 1000000  # Ejemplo: asignar el rea como nuevo valor

            # Actualizar los atributos con el nuevo campo
            feature['properties']['Area km2'] = areakm

            # Escribir la geometra actualizada en el nuevo shapefile
            output.write(feature)
```

### Reproject

```python
import fiona
from fiona.transform import transform_geom

input_path = '/home/diego/git/AEPython_2024/data/13_01_TerminoMunicipal.shp'
output_path = '/home/diego/git/AEPython_2024/data/13_01_TerminoMunicipal_gofio.shp'

# Sistema de referencia de destino (EPSG:4326, WGS84)
target_crs = 'EPSG:4083'

# Leer el shapefile original
with fiona.open(input_path) as shp:
    schema = shp.schema
    
    # Crear un nuevo shapefile reproyectado
    with fiona.open(output_path, 'w', driver='ESRI Shapefile', schema=schema, crs=target_crs) as output:
        for feature in shp:
            # Reproyectar la geometra
            transformed_geometry = transform_geom(
                shp.crs, target_crs, feature['geometry']
            )

            # Escribir la geometra reproyectada en el nuevo shapefile
            output.write({
                'geometry': transformed_geometry,
                'properties': feature['properties']
            })
```
